# Notebook for cleaning `ENCOUNTERS TABLE`

```python
import pandas as pd
import numpy as np
 
from sqlalchemy import create_engine,text
user = 'root'
passwords = '7003890541'
host = 'localhost'
port = 3306
database = 'synthea'

engine = create_engine(f"mysql+pymysql://{user}:{passwords}@{host}:{port}/{database}")
def sql(query , engine = engine):
    return pd.read_sql(query,engine)
```

```python
def select(table_name):
    return sql(f'select * from {table_name} limit 5')
```

```python
def run(query):
    with engine.connect() as con:
        con.execute(text(query))
        con.commit()
```

```python
sql('select * from encounters limit 5')
```

```python
query = '''
    select column_name from information_schema.columns
    where table_name = 'encounters' and
    table_schema = database();
'''
sql(query)
encounters_columns = [c[0] for c in sql(query).values.tolist()]
```

```python
for col in encounters_columns:
    query = f'''
        alter table encounters
        rename column {col} to {col.lower()}
    '''
    with engine.connect() as con:
        con.execute(text(query))
        con.commit()
```

```python
for col in encounters_columns:
    query = f'''
        select sum({col} is null) as {col}_null from encounters
        having {col}_null > 0;
    '''
    display(sql(query))
```
